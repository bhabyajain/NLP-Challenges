```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

hrnlp1=open("training.txt","r")
text=hrnlp1.read()
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
stop=list(stopwords.words('english'))
import string
punctuations=list(string.punctuation)
slist=[]
check=0
sclass=[]
temp=[]
nol=''
for i in range(len(text)):
    #print(temp)
    if(check==0 and text[i]=='\n'):
        check=1
        temp1=''.join(temp)
        nol=temp1
        temp=[]
    elif(check==1 and text[i]=='\t'):
        check=2
        temp1=''.join(temp)
        slist.append(temp1)
        temp=[]
    elif(check==2 and text[i]=='\n'):
        check=1
        temp1=''.join(temp)
        sclass.append(temp1)
        temp=[]
    elif(i+1==len(text)):
        temp.append(text[i])
        temp1=''.join(temp)
        sclass.append(temp1)
        temp=[]
    else:
        temp.append(text[i])
def removepunc(t):
    t1 = ''.join(ch for ch in t if ch not in punctuations)
    import re
    t2 = re.sub(' +', ' ', t1)
    return t2
def removestop(t):
    t1=t.split(" ")
    t2=""
    for i in range(len(t1)):
        if(t1[i].lower() not in stop):
            #allwords.append(t1[i].lower())
            t2=t2+t1[i].lower()
            t2=t2+" "
    return t2
for i in range(len(slist)):
    slist[i]=removepunc(slist[i])
    slist[i]=removestop(slist[i])
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ytrain=le.fit_transform(sclass)
ssclass=sclass
features=[]
for i in range(len(ssclass)):
    part=ssclass[i].split(" ")
    for j in range(len(part)):
        ppart=part[j].split("-")
        for k in range(len(ppart)):
            if(ppart[k].lower() not in features):
                features.append(ppart[k].lower())
xtrain=np.zeros((len(slist),len(features)))
for i in range(len(slist)):
    temp=slist[i].split(" ")
    for j in range(len(temp)):
        if(temp[j] in features):
            idd=features.index(temp[j])
            #if(i>=70):
                #print(slist[i])
                #print(temp[j],idd,i,sclass[i])
            xtrain[i,idd]=xtrain[i,idd]+1
n=int(input())
tlist=[]
for i in range(n):
    ss=input()
    rp=removepunc(ss)
    rs=removestop(rp)
    tlist.append(rs)
xtest=np.zeros((len(tlist),len(features)))
for i in range(len(tlist)):
    temp=tlist[i].split(" ")
    for j in range(len(temp)):
        if(temp[j] in features):
            idd=features.index(temp[j])
            xtest[i,idd]=xtest[i,idd]+1
pred=[]
for i in range(len(xtest)):
    maxprob=0
    maxprobclass=5
    temp=xtest[i]
    for j in range(20):
        prob=1
        #ccount=0
        for k in range(len(temp)):
            tsum=0
            csum=0
            if(temp[k]!=0):
                for l in range(len(xtrain)):
                    if(ytrain[l]==j):
                        #ccount=ccount+1
                        temp1=xtrain[l]
                        for m in range(len(temp1)):
                            tsum=tsum+temp1[m]
                            if(m==k):
                                csum=csum+temp1[m]
                prob=(prob)*(csum+1/tsum)
        #prob=prob*ccount
        if(prob>maxprob):
            maxprob=prob
            maxprobclass=j
    pred.append(maxprobclass)
ans=le.inverse_transform(pred)
print(ans)
```
