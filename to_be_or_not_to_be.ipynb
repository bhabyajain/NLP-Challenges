```python
import nltk
from nltk import pos_tag
#nltk.download('averaged_perceptron_tagger')
n=int(input())
slist=[]
k=0
while(k<n):
    s=input()
    temp1=s.split(".")
    for i in range(len(temp1)):
        temp2=temp1[i].split("----")
        if(len(temp2) > 1):
            slist.append(temp2)
        k=k+len(temp2)-1
punctuations=[',']
def removepunc(t):
    t1 = ''.join(ch for ch in t if ch not in punctuations)
    import re
    t2 = re.sub(' +', ' ', t1)
    if(t2[0]==' '):
        t2=t2[1:]
    if(t2[len(t2)-1]==' '):
        t2=t2[0:len(t2)-1]
    return t2
sslist=[]
for i in range(len(slist)):
    temp=slist[i]
    tt=[]
    for j in range(len(temp)):
        temp1=removepunc(temp[j])
        tt.append(temp1)
    sslist.append(tt)

def findtense(textarr):
    tense = [0,0]
    for i in range(len(textarr)):
        words=textarr[i].split(" ")
        tagged = pos_tag(words)
        #tense[2] = tense[2]+len([word for word in tagged if word[1] == "MD"])
        tense[1] = tense[1]+len([word for word in tagged if word[1] in ["VBP", "VBZ","VBG"]])
        tense[0] = tense[0]+len([word for word in tagged if word[1] in ["VBD", "VBN"]]) 
    if(tense[0]!=0):
        return 0
    else:
        return 1
    return(tense)
#print ("here")
def findsingularity(text):
    c1=0
    temp=-1
    words=text.split(" ")
    #print(words)
    for i in range(len(words)-1,-1,-1):
        #print("here")
        tw=[words[i]]
        sp=pos_tag(tw)
        pos=sp[0]
        kk=pos[1]
        #print(tw,kk)
        if(c1==0 and kk[0]=='N'):
            if(len(kk)>2 and kk[len(kk)-1]=='S'):
                return 1
            else:
                return 0
        if(c1==0 and (words[i]=="He" or words[i]=="She" or words[i]=="he" or words[i]=="she" or words[i]=="I" or words[i]=="i")):
            return 0
        if(c1==0 and (words[i]=="They" or words[i]=="We" or words[i]=="they" or words[i]=="we" or words[i]=="us" or words[i]=="Us")):
            return 1
        if(c1==0 and kk[0]=='V'):
            c1=1
        if(c1==1 and kk[0]=='N' and temp==-1):
            if(len(kk)>2 and kk[len(kk)-1]=='S'):
                temp=1
            else:
                temp=0
        if(c1==1 and (words[i]=="He" or words[i]=="She" or words[i]=="he" or words[i]=="she" or words[i]=="I" or words[i]=="i") and temp==-1):
            temp=0
        if(c1==1 and (words[i]=="They" or words[i]=="We" or words[i]=="they" or words[i]=="we" or words[i]=="us" or words[i]=="Us") and temp==-1):
            temp=1
def displaypr(textarr):
    for i in range(len(textarr)-1):
        to=textarr[i]
        #print(to)
        wrd=to.split(" ")
        #print(wrd)
        if(wrd[len(wrd)-1]=='has' or wrd[len(wrd)-1]=='have' or wrd[len(wrd)-1]=='had' or wrd[len(wrd)-1]=='Has' or wrd[len(wrd)-1]=='Have' or wrd[len(wrd)-1]=='Had'):
            print("been")
            continue
        #print("here")
        psp=[wrd[len(wrd)-1]]
        #print("here")
        spp=pos_tag(psp)
        #print("here")
        sp1=spp[0]
        if(wrd[len(wrd)-1]=='to' or wrd[len(wrd)-1]=='To' or sp1[1]=="MD"):
            print("to")
            continue
        #print("here")
        tns=findtense(textarr)
        sg=findsingularity(to)
        if(wrd[len(wrd)-1]=='I' or wrd[len(wrd)-1]=='i'):
            if(tns==1):
                print("am")
                continue
            else:
                print("was")
                continue
        else:
            if(tns==1 and sg==1):
                print("are")
                continue
            if(tns==1 and sg==0):
                print("is")
                continue
            if(tns==0 and sg==1):
                print("were")
                continue
            if(tns==0 and sg==0):
                print("was")
                continue
for i in range(len(sslist)):
    #print(sslist[i])
    displaypr(sslist[i])
```
